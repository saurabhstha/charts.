```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
```

```python
#load dataset
df = pd.read_csv('expenditures.txt', sep="\t")
```

```python
#view data
df
```

```python
# Calculate total expenditure for categories
expen_cat = df.groupby(['category'])['expenditure'].sum().reset_index()


#Create our plot and resize it.
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(14, 8)

# Plot tree map now
squarify.plot(sizes=expen_cat['expenditure'], 
              label=expen_cat['category'], 
              alpha=.8 )
plt.title('Expenditure by Category')
plt.axis('off')
plt.show()
```

```python
#create pivot table
df1= df.pivot(index='year', columns='category', values= 'expenditure')
df1
```

```python
#create stacked area chart using above pivot table

df1.plot.area(figsize = (17,8))
plt.title("Expenditure by year\n", fontsize = 5, loc = 'right')
plt.ylabel("Expenditure", fontsize = 9)
plt.xlabel("Year", fontsize = 9)
plt.show()
```

```python
# Calculate total expenditure by year
exp_year = df.groupby(['year'])['expenditure'].sum().reset_index()

# Area Chart
plt.fill_between(exp_year['year'], exp_year['expenditure'])
plt.xlabel('Year')
plt.ylabel('Expenditure')
plt.title('Expenditure by Year')
plt.show()
```
